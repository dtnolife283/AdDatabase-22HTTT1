### order
> AE ĐỂ ID_CUSTOMER NULL NHA, LẤY branch_table_list, để orderDate vào năm 2022, 2023, để totalPrice với actualPrice = 0 hết nha

ORDER
- ID_Order INT,
- ID_Table INT,
- OrderDate DATE NOT NULL DEFAULT GETDATE(),
- TotalPrice INT,
- ActualPrice INT,
- ID_Customer INT,
- ID_Review INT,
- ID_Employee INT,
- ID_Branch INT,
- CONSTRAINT PK_ID_ORDER PRIMARY KEY (ID_Order)

```python
def generate_table_data(branch_id_list):
    branch_table_list = {}
    id_table_count = 1
    for branch_id in branch_id_list:
        branch_table_list[branch_id] = []
        num_records = fake.random_int(20, 30)
        for i in range(1, num_records + 1):
            id_table = id_table_count
            branch_table_list[branch_id].append(id_table)
            number_of_seats = fake.random_element([2, 4, 6, 8])
            status_table = random.choices([0, 1], weights=[0.5, 0.5], k=1)[0]
            # print(
            #     f"INSERT INTO TABLE (ID_Table, NumberOfSeats, Status, ID_Branch) VALUES ({id_table}, {number_of_seats}, {status_table}, {branch_id});"
            # )
            id_table_count += 1
    return branch_table_list

branch_id_list = list(range(1, no_of_branches + 1))
branch_table_list = generate_table_data(branch_id_list)
```

```python
def generate_order_id_list(branch_table_list: dict, no_records: int, has_table_ratio: float = 0.8) -> dict:

    order_id_list = {
        "ID_Order": [],
        "ID_Table": [],
        "OrderDate": [],
        "TotalPrice": [],
        "ActualPrice": [],
        "ID_Customer": [],
        "ID_Review": [],
        "ID_Employee": [],
        "ID_Branch": [],
    }
    TotalPrice, ActualPrice = 0, 0
    review_set = set()
    
    print("INSERT INTO ORDER\nVALUES")
    
    for ID_Branch, tableIDs in branch_table_list.items():
        for i in range(1, no_records + 1):
            ID_Order = i
            ID_Table = random.choice(tableIDs) if i < (no_records * has_table_ratio) else "NULL" # 60% of orders are from tables
            OrderDate = fake.date_between(start_date = "-2y", end_date = "-1y")
            ID_Customer = random.randint(1, no_records + 1)
    
            ID_Review = random.randint(1, no_records + 1)
            while ID_Review in review_set:
                ID_Review = random.randint(1, no_records + 1)
            review_set.add(ID_Review)
            
            ID_Employee = random.randint(1, no_of_employees + 1)

            print(
                f"\t({ID_Order}, {ID_Table}, {OrderDate}, {TotalPrice}, {ActualPrice}, {ID_Customer}, {ID_Review}, {ID_Employee}, {ID_Branch})",
                end = ",\n"
            )
            order_id_list["ID_Order"].append(ID_Order)
            order_id_list["ID_Table"].append(ID_Table)
            order_id_list["OrderDate"].append(OrderDate)
            order_id_list["TotalPrice"].append(TotalPrice)
            order_id_list["ActualPrice"].append(ActualPrice)
            order_id_list["ID_Customer"].append(ID_Customer)
            order_id_list["ID_Review"].append(ID_Review)
            order_id_list["ID_Employee"].append(ID_Employee)
            order_id_list["ID_Branch"].append(ID_Branch)

        review_set.clear()

    return order_id_list

order_id_list = generate_order_id_list(branch_table_list, 20)
```

### order-food
> AE viết cái procedure thêm món đã - rồi gọi nó ở đây cho nó cập nhật cái order ở trên nha

ORDER_FOOD
- ID_BranchFood INT
- ID_Order INT,
- Quantity INT NOT NULL CHECK (Quantity > 0),
- CONSTRAINT PK_ORDER_FOOD PRIMARY KEY (ID_BranchFood, ID_Order)


```python
def generate_order_food_data(branch_foods: dict, order_id_list: dict, foodID_list: list):
    # branch_foods[(id_branch, id_food)] = id_branchfood

    # before inserting, check if:
    # - ID_Order in ORDER_FOOD exists in ORDER
    # - ID_BranchFood in ORDER_FOOD exists in BRANCH_FOOD
    
    print("""
        CREATE PROC INSERT_ORDER_FOOD
            @ID_BranchFood INT,
            @ID_Order INT,
            @Quantity INT
        AS
            BEGIN
                --check if quantity = 0, if so do not insert
                IF @Quantity = 0
                BEGIN
                    PRINT 'Quantity cannot be 0';
                    RETURN;
                END;

                IF  EXISTS (SELECT ID_Order FROM ORDER WHERE ID_Order = @ID_Order)
                AND EXISTS (SELECT ID_BranchFood FROM BRANCH_FOOD WHERE ID_BranchFood = @ID_BranchFood)
                BEGIN
                    INSERT INTO ORDER_FOOD VALUES (@ID_Order, @ID_BranchFood, @Quantity);
                END;
            END;
        BEGIN
        END;
        """)
    
    for orderID, branchID in zip(order_id_list["ID_Order"], order_id_list["ID_Branch"]):
        no_food = random.randint(1, 10)
        for j in range(no_food):
            while True:
                foodID = random.choice(foodID_list)
                if (branchID, foodID) in branch_foods:
                    ID_BranchFood = branch_foods[(branchID, foodID)]
                    break
            Quantity = random.randint(1, 5)
            # call the stored procedure
            print(f"EXEC INSERT_ORDER_FOOD {ID_BranchFood}, {orderID}, {Quantity};")
```

### online-order

- ID_Online INT,
- TimeOrder TIME NOT NULL,
- CONSTRAINT PK_ONLINE_ORDER PRIMARY KEY (ID_Online)

```python
index = order_id_list["ID_Table"].index("NULL")
reservation_ratio = random.randint(1, 3) * 0.1


reservation_orders = {key: value[int(index * reservation_ratio):] for key, value in order_id_list.items()} # table = NULL => reservation order
online_orders = {key: value[:index] for key, value in order_id_list.items()} # table = NULL => online order



def generate_online_order(online_order):
    for orderID in online_order["ID_Order"]:
        TimeOrder = fake.time(pattern="%H:%M:%S")
        print(
            f"INSERT INTO ONLINE_ORDER VALUES ({orderID}, {TimeOrder});"
        )

generate_online_order(online_orders)
```

### order
> AE ĐỂ ID_CUSTOMER NULL NHA, LẤY branch_table_list, để orderDate vào năm 2022, 2023, để totalPrice với actualPrice = 0 hết nha

ORDER
- ID_Order INT,
- OrderDate DATE NOT NULL DEFAULT GETDATE(),
- TotalPrice INT,
- ActualPrice INT,
- ID_Customer INT,
- ID_Review INT,
- ID_Employee INT,
- ID_Branch INT,
- CONSTRAINT PK_ID_ORDER PRIMARY KEY (ID_Order)

```python


def generate_order_id_list(no_records):
    
    order_id_list = []
    order_date = [2022, 2023]
    totalPrice, actualPrice = 0, 0
    review_set = set()
    
    for i in range(1, no_records + 1):
        ID_Order = i

        OrderDate = fake.date_between(start_date="-2y", end_date="-1y") # 2022, 2023
        TotalPrice = 0
        ActualPrice = 0
        ID_Customer = i if i <= no_records * 0.7 else "NULL"
        
        ID_review = random.randint(1, no_records)
        while ID_review in review_set:
            ID_review = random.randint(1, no_records)
        review_set.add(ID_review)
        
        ID_Employee = random.randint(1, no_of_employees)
        ID_Branch = random.randint(1, no_of_branches)

        order_id_list.append((ID_Order, ID_Branch))
        print(
            f"INSERT INTO [ORDER] VALUES ({ID_Order}, '{OrderDate}', {TotalPrice}, {ActualPrice}, {ID_Customer}, {ID_review}, {ID_Employee}, {ID_Branch});"
        )

    return order_id_list

order_id_list = generate_order_id_list(1000)
```
